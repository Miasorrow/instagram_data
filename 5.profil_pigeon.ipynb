```python
df = pd.read_csv("./dataset/instagram_users_lifestyle.csv")

df.head()        # voir les premi√®res lignes
df.info()        # types des colonnes
df.describe()    # stats g√©n√©rales
```

```python
import pandas as pd

# üîπ Cr√©er quartiles sur les clics publicitaires
df['ads_click_quartile'] = pd.qcut(
    df['ads_clicked_per_day'],
    4,
    labels=['Q1_Low', 'Q2_Medium', 'Q3_High', 'Q4_VeryHigh']
)

# üîπ Colonnes num√©riques
num_columns = [
    'age', 'exercise_hours_per_week', 'sleep_hours_per_night',
    'perceived_stress_score', 'self_reported_happiness',
    'body_mass_index', 'blood_pressure_systolic', 'blood_pressure_diastolic'
]

# üîπ Colonnes cat√©gorielles
cat_columns = ['income_level']  # on peut ajouter d'autres plus tard

# üîπ Moyennes pour les num√©riques
num_summary = df.groupby('ads_click_quartile')[num_columns].mean().round(2)

# üîπ Mode pour les cat√©gorielles
cat_summary = df.groupby('ads_click_quartile')[cat_columns].agg(lambda x: x.mode().iloc[0])

# üîπ Combiner num√©rique + cat√©goriel
summary = pd.concat([num_summary, cat_summary], axis=1)

# üîπ Cr√©er tableau ‚Äúligne = Clic le plus / Clic le moins / Diff√©rence‚Äù
table = pd.DataFrame({
    'Clic_le_plus': summary.loc['Q4_VeryHigh'],
    'Clic_le_moins': summary.loc['Q1_Low']
})

# üîπ Pour les num√©riques, calculer la diff√©rence
diff_numeric = table.loc[num_columns, 'Clic_le_plus'] - table.loc[num_columns, 'Clic_le_moins']
table.loc['Diff√©rence'] = diff_numeric

# üîπ Pour les colonnes cat√©gorielles, on met juste ‚Äú‚Äî‚Äù pour diff√©rence
for col in cat_columns:
    table.loc['Diff√©rence', col] = '‚Äî'

# üîπ Transposer pour affichage lignes = Clic le plus / Clic le moins / Diff√©rence
table_transposed = table.T

# üîπ Afficher tout correctement
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.float_format', '{:.2f}'.format)

table_transposed
```

```python
# üîπ Colonnes cat√©gorielles √† compter
cat_columns = ['income_level', 'subscription_status', 'relationship_status']

# üîπ Cr√©er un tableau final vide
final_table = pd.DataFrame()

for quartile in ['Clic_le_moins', 'Clic_le_plus']:
    # S√©lection du quartile
    df_q = df[df['ads_click_quartile'] == quartile]
    
    # Num√©riques ‚Üí moyenne
    num_mean = df_q[num_columns].mean().round(2)
    
    # Cat√©gorielles ‚Üí distribution (%) de chaque cat√©gorie
    cat_counts = {}
    for col in cat_columns:
        counts = df_q[col].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
        # On met les valeurs dans une cha√Æne genre "30k-50k:40%,50k-70k:60%"
        cat_counts[col] = ', '.join([f"{i}: {v}" for i,v in zip(counts.index, counts.values)])
    
    # Combiner num√©riques + cat√©gorielles
    row = pd.concat([num_mean, pd.Series(cat_counts)])
    final_table = pd.concat([final_table, row.to_frame().T], ignore_index=True)

# Renommer les lignes
final_table.index = ['Clic_le_moins', 'Clic_le_plus']

# Moyenne des num√©riques
mean_row = final_table[num_columns].astype(float).mean().round(2)
# Pour les cat√©gorielles, on peut mettre ‚Äú‚Äî‚Äù ou choisir le plus fr√©quent globalement
for col in cat_columns:
    mean_row[col] = '‚Äî'
final_table.loc['Moyenne'] = mean_row

final_table
```

```python
import pandas as pd
import numpy as np

# üîπ Cr√©er quartiles
df['ads_click_quartile'] = pd.qcut(
    df['ads_clicked_per_day'],
    4,
    labels=['Clic_le_moins', 'Q2', 'Q3', 'Clic_le_plus']
)

# üîπ Colonnes
num_columns = [
    'age', 'exercise_hours_per_week', 'sleep_hours_per_night',
    'perceived_stress_score', 'self_reported_happiness',
    'body_mass_index'
]

cat_columns = [
    'income_level',
    'subscription_status',
    'relationship_status'
]

# ==========================
# üîπ PARTIE NUM√âRIQUE
# ==========================

num_summary = df.groupby('ads_click_quartile')[num_columns].mean().round(2)

# ==========================
# üîπ PARTIE CAT√âGORIELLE
# ==========================

cat_summary = {}

for quartile in ['Clic_le_moins', 'Clic_le_plus']:
    df_q = df[df['ads_click_quartile'] == quartile]
    cat_summary[quartile] = {}
    
    for col in cat_columns:
        value = df_q[col].value_counts().idxmax()
        count = df_q[col].value_counts().max()
        cat_summary[quartile][col] = f"{value} ({count})"

cat_summary_df = pd.DataFrame(cat_summary).T

# ==========================
# üîπ COMBINAISON
# ==========================

table = pd.concat([
    num_summary.loc[['Clic_le_moins','Clic_le_plus']],
    cat_summary_df
], axis=1)

# ==========================
# üîπ DIFF√âRENCE NUM√âRIQUE
# ==========================

difference = (
    num_summary.loc['Clic_le_plus'] -
    num_summary.loc['Clic_le_moins']
).round(2)

# Ajouter diff√©rence
diff_row = pd.Series(index=table.columns)
diff_row[num_columns] = difference
diff_row[cat_columns] = "‚Äî"

table.loc['Diff√©rence'] = diff_row

# ==========================
# üîπ AFFICHAGE COMPLET
# ==========================

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

table
```
