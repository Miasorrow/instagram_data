On va dÃ©couper posts_created_per_week en 4 groupes :

Q1 â†’ petits crÃ©ateurs

Q2 â†’ moyens

Q3 â†’ gros

Q4 â†’ trÃ¨s gros crÃ©ateurs

```python
import pandas as pd

# CrÃ©ation des quartiles
df['creator_quartile'] = pd.qcut(
    df['posts_created_per_week'],
    3,
    labels=['Q1_Petit', 'Q2_Moyen', 'Q3_Gros']
)

# VÃ©rification
print(df['creator_quartile'].value_counts())
```

```python
import pandas as pd
import numpy as np

# ðŸ”¹ Colonnes pertinentes pour l'analyse du comportement
columns_to_use = [
    # ActivitÃ© / bien-Ãªtre
    'exercise_hours_per_week', 'sleep_hours_per_night', 'body_mass_index',
    'perceived_stress_score', 'self_reported_happiness',
    
    # Engagement gÃ©nÃ©ral / social
    'daily_steps_count', 'weekly_work_hours', 'hobbies_count',
    'social_events_per_month', 'books_read_per_year', 'volunteer_hours_per_month',
    'travel_frequency_per_year',
    
    # ActivitÃ© sur l'application
    'daily_active_minutes_instagram', 'sessions_per_day', 'posts_created_per_week',
    'reels_watched_per_day', 'stories_viewed_per_day', 'likes_given_per_day',
    'comments_written_per_day', 'dms_sent_per_week', 'dms_received_per_week',
    'ads_viewed_per_day', 'ads_clicked_per_day', 'time_on_feed_per_day',
    'time_on_explore_per_day', 'time_on_messages_per_day', 'time_on_reels_per_day',
    'following_count', 'notification_response_rate',
    'average_session_length_minutes', 'user_engagement_score'
]

# ðŸ”¹ Nettoyer les donnÃ©es pour posts_created_per_week (quartiles)
df_clean = df.dropna(subset=['posts_created_per_week'])

# ðŸ”¹ CrÃ©ation des quartiles Q1 Ã  Q4
df_clean['creator_quartile'] = pd.qcut(
    df_clean['posts_created_per_week'],
    4,
    labels=['Q1_Petit', 'Q2_Bas', 'Q3_Haut', 'Q4_Gros']
)

# ðŸ”¹ SÃ©lection des colonnes numÃ©riques pertinentes
df_numeric = df_clean[columns_to_use]

# ðŸ”¹ Moyennes par quartile
grouped = df_numeric.groupby(df_clean['creator_quartile']).mean()

# ðŸ”¹ Calcul Ã©cart Q4 - Q1
diff = grouped.loc['Q4_Gros'] - grouped.loc['Q1_Petit']

# ðŸ”¹ CrÃ©ation du tableau final
result = pd.DataFrame({
    'Q1_Petit': grouped.loc['Q1_Petit'],
    'Q4_Gros': grouped.loc['Q4_Gros'],
    'Diff_Q4_minus_Q1': diff
})

# ðŸ”¹ Arrondir
result = result.round(2)

# ðŸ”¹ Trier du plus grand Ã©cart au plus petit
result = result.sort_values(by='Diff_Q4_minus_Q1', ascending=False)

print(result)
```
